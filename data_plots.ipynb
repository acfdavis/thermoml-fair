```python
import pandas as pd
import matplotlib.pyplot as plt

# === Step 1: Load your dataset ===
# Replace with your actual file and column name
df = pd.read_csv("thermoml_data.csv")  # or .json/.parquet depending on your output
property_col = "property"  # adjust if your column is named differently

# === Step 2: Count occurrences of each property ===
property_counts = df[property_col].value_counts()

# === Step 3: Select Top 10 ===
top10 = property_counts.head(10)

# === Step 4: Plot ===
plt.figure(figsize=(10,6))
bars = plt.barh(top10.index[::-1], top10.values[::-1])  # horizontal bar chart

# Add labels on each bar
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.01*max(top10.values),
             bar.get_y() + bar.get_height()/2,
             f"{width:,}", va="center")

# Titles and labels
plt.title("Top 10 Properties Extracted from ThermoML Dataset", fontsize=14, weight='bold')
plt.xlabel("Number of Entries")
plt.ylabel("Property")

# Tight layout for clean spacing
plt.tight_layout()

# Save chart
plt.savefig("thermoml_top10_properties.png", dpi=300)
plt.show()
```
